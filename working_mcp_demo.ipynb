## Setup and Database Connection

First, let's set up the MCP tools and connect to your simulation database.

```python
# Import required modules
import sys
import json
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

# Add the agentfarm_mcp module to the path
sys.path.append('/workspace')

print("=== AgentFarm MCP Real Data Analysis Demo ===")
print()

# Import MCP components
from agentfarm_mcp.config import MCPConfig
from agentfarm_mcp.services.database_service import DatabaseService
from agentfarm_mcp.services.cache_service import CacheService

# Import the tools we'll use
from agentfarm_mcp.tools.metadata_tools import ListSimulationsTool, GetSimulationInfoTool
from agentfarm_mcp.tools.query_tools import (
    QueryAgentsTool, QueryActionsTool, QueryStatesTool, 
    QueryResourcesTool, GetSimulationMetricsTool
)
from agentfarm_mcp.tools.analysis_tools import (
    AnalyzePopulationDynamicsTool, AnalyzeSurvivalRatesTool,
    AnalyzeResourceEfficiencyTool, IdentifyCriticalEventsTool
)
from agentfarm_mcp.tools.advanced_tools import BuildAgentLineageTool, GetAgentLifecycleTool

print("Successfully imported all MCP tools!")
print()

# Initialize database and cache services
config = MCPConfig.from_db_path('/workspace/simulation.db')
db_service = DatabaseService(config.database)
cache_service = CacheService(config.cache)

print(f"Connected to database: {config.database.path}")
print(f"Database type: {config.database.database_type}")
print(f"Cache enabled: {config.cache.enabled}")
print()

# Initialize tools
list_simulations_tool = ListSimulationsTool(db_service, cache_service)
get_simulation_info_tool = GetSimulationInfoTool(db_service, cache_service)
query_agents_tool = QueryAgentsTool(db_service, cache_service)
query_actions_tool = QueryActionsTool(db_service, cache_service)
query_states_tool = QueryStatesTool(db_service, cache_service)
query_resources_tool = QueryResourcesTool(db_service, cache_service)
get_simulation_metrics_tool = GetSimulationMetricsTool(db_service, cache_service)
analyze_population_tool = AnalyzePopulationDynamicsTool(db_service, cache_service)
analyze_survival_tool = AnalyzeSurvivalRatesTool(db_service, cache_service)
analyze_resource_tool = AnalyzeResourceEfficiencyTool(db_service, cache_service)
identify_events_tool = IdentifyCriticalEventsTool(db_service, cache_service)
build_lineage_tool = BuildAgentLineageTool(db_service, cache_service)
get_lifecycle_tool = GetAgentLifecycleTool(db_service, cache_service)

print("All tools initialized successfully!")
print()

# Get available simulations
print("=== Available Simulations ===")
try:
    simulations_result = list_simulations_tool.execute(
        status=None,
        experiment_id=None,
        limit=100,
        offset=0
    )
    print(f"Found {len(simulations_result.get('simulations', []))} simulations:")
    
    # Store the first simulation ID for our analysis
    if simulations_result.get('simulations'):
        simulation_id = simulations_result['simulations'][0]['simulation_id']
        print(f"Using simulation ID: {simulation_id}")
        
        # Get detailed simulation info
        sim_info = get_simulation_info_tool.execute(simulation_id=simulation_id)
        print(f"Simulation status: {sim_info.get('status', 'Unknown')}")
        print(f"Start time: {sim_info.get('start_time', 'Unknown')}")
        print(f"End time: {sim_info.get('end_time', 'Still running')}")
    else:
        print("No simulations found in database")
        simulation_id = None
        
except Exception as e:
    print(f"Error getting simulations: {e}")
    simulation_id = None

print()
```

## Basic Data Queries

Let's start with some basic queries to understand the data structure.

```python
if simulation_id:
    print("=== Basic Data Queries ===")
    print()
    
    # Query agents (without offset to avoid the error)
    print("1. Querying Agents:")
    try:
        agents_result = query_agents_tool.execute(
            simulation_id=simulation_id,
            limit=10
        )
        print(f"Total agents: {agents_result.get('total_count', 0)}")
        print(f"Returned: {agents_result.get('returned_count', 0)} agents")
        
        if agents_result.get('agents'):
            print("Sample agents:")
            for i, agent in enumerate(agents_result['agents'][:3]):
                print(f"  Agent {i+1}: {agent['agent_id']} (type: {agent['agent_type']}, gen: {agent['generation']})")
    except Exception as e:
        print(f"Error querying agents: {e}")
    print()
    
    # Query simulation metrics (without offset)
    print("2. Querying Simulation Metrics:")
    try:
        metrics_result = get_simulation_metrics_tool.execute(
            simulation_id=simulation_id,
            limit=20
        )
        print(f"Total metric records: {metrics_result.get('total_count', 0)}")
        print(f"Returned: {metrics_result.get('returned_count', 0)} records")
        
        if metrics_result.get('metrics'):
            print("Sample metrics:")
            for i, metric in enumerate(metrics_result['metrics'][:3]):
                print(f"  Step {metric['step_number']}: {metric['total_agents']} agents, {metric['total_resources']} resources")
    except Exception as e:
        print(f"Error querying metrics: {e}")
    print()
    
    # Query actions (without offset)
    print("3. Querying Actions:")
    try:
        actions_result = query_actions_tool.execute(
            simulation_id=simulation_id,
            limit=10
        )
        print(f"Total actions: {actions_result.get('total_count', 0)}")
        print(f"Returned: {actions_result.get('returned_count', 0)} actions")
        
        if actions_result.get('actions'):
            print("Sample actions:")
            for i, action in enumerate(actions_result['actions'][:3]):
                print(f"  Action {i+1}: {action['action_type']} by {action['agent_id']} at step {action['step_number']}")
    except Exception as e:
        print(f"Error querying actions: {e}")
    print()
    
    # Query resources (without offset)
    print("4. Querying Resources:")
    try:
        resources_result = query_resources_tool.execute(
            simulation_id=simulation_id,
            limit=10
        )
        print(f"Total resource records: {resources_result.get('total_count', 0)}")
        print(f"Returned: {resources_result.get('returned_count', 0)} records")
        
        if resources_result.get('resources'):
            print("Sample resources:")
            for i, resource in enumerate(resources_result['resources'][:3]):
                print(f"  Resource {i+1}: {resource['resource_id']} at ({resource['position']['x']:.1f}, {resource['position']['y']:.1f}) with {resource['amount']} units")
    except Exception as e:
        print(f"Error querying resources: {e}")
    print()
else:
    print("No simulation ID available for queries")
```

## Population Analysis

Now let's analyze the population dynamics using the real data.

```python
if simulation_id:
    print("=== Population Analysis ===")
    print()
    
    # Analyze population dynamics
    print("1. Population Dynamics Analysis:")
    try:
        pop_result = analyze_population_tool.execute(
            simulation_id=simulation_id,
            include_chart=True
        )
        
        if 'error' not in pop_result:
            summary = pop_result.get('population_summary', {})
            print(f"Initial population: {summary.get('initial_population', 'N/A')}")
            print(f"Final population: {summary.get('final_population', 'N/A')}")
            print(f"Peak population: {summary.get('peak_population', 'N/A')} at step {summary.get('peak_step', 'N/A')}")
            print(f"Growth rate: {summary.get('total_growth_rate_percent', 'N/A')}%")
            print(f"Total births: {summary.get('total_births', 'N/A')}")
            print(f"Total deaths: {summary.get('total_deaths', 'N/A')}")
            print(f"Net change: {summary.get('net_change', 'N/A')}")
            
            # Show the ASCII chart if available
            if 'chart' in pop_result:
                print("\nPopulation Chart:")
                print(pop_result['chart'])
        else:
            print(f"Error in population analysis: {pop_result['error']}")
            
    except Exception as e:
        print(f"Error analyzing population: {e}")
    print()
    
    # Analyze survival rates
    print("2. Survival Rates Analysis:")
    try:
        survival_result = analyze_survival_tool.execute(
            simulation_id=simulation_id,
            group_by='generation'
        )
        
        if 'error' not in survival_result:
            summary = survival_result.get('summary', {})
            print(f"Overall survival rate: {summary.get('overall_survival_rate', 'N/A')}%")
            print(f"Total agents: {summary.get('total_agents', 'N/A')}")
            print(f"Number of generations: {summary.get('total_groups', 'N/A')}")
            
            print("\nSurvival by generation:")
            cohorts = survival_result.get('cohorts', {})
            for gen, data in cohorts.items():
                print(f"  Generation {gen}: {data.get('survival_rate_percent', 'N/A')}% survival ({data.get('alive', 0)}/{data.get('total_agents', 0)})")
        else:
            print(f"Error in survival analysis: {survival_result['error']}")
            
    except Exception as e:
        print(f"Error analyzing survival: {e}")
    print()
    
    # Identify critical events
    print("3. Critical Events Analysis:")
    try:
        events_result = identify_events_tool.execute(
            simulation_id=simulation_id,
            threshold_percent=10.0
        )
        
        if 'events' in events_result:
            events = events_result['events']
            print(f"Found {len(events)} critical events:")
            
            for i, event in enumerate(events[:5]):  # Show first 5 events
                print(f"  Event {i+1}: {event['type']} at step {event['step']} - {event['description']} (severity: {event['severity']})")
            
            if len(events) > 5:
                print(f"  ... and {len(events) - 5} more events")
                
            # Show summary
            summary = events_result.get('summary', {})
            print(f"\nEvent summary:")
            print(f"  Total events: {summary.get('total_events', 0)}")
            
            by_type = summary.get('by_type', {})
            if by_type:
                print(f"  By type: {by_type}")
                
            by_severity = summary.get('by_severity', {})
            if by_severity:
                print(f"  By severity: {by_severity}")
        else:
            print(f"Error in events analysis: {events_result.get('error', 'Unknown error')}")
            
    except Exception as e:
        print(f"Error analyzing events: {e}")
    print()
else:
    print("No simulation ID available for population analysis")
```

## Agent Analysis

Let's analyze individual agents and their lifecycles.

```python
if simulation_id:
    print("=== Agent Analysis ===")
    print()
    
    # Get a sample agent for detailed analysis
    try:
        agents_result = query_agents_tool.execute(
            simulation_id=simulation_id,
            limit=5
        )
        
        if agents_result.get('agents'):
            sample_agent = agents_result['agents'][0]
            agent_id = sample_agent['agent_id']
            
            print(f"Analyzing agent: {agent_id}")
            print(f"Type: {sample_agent['agent_type']}")
            print(f"Generation: {sample_agent['generation']}")
            print(f"Birth time: {sample_agent['birth_time']}")
            print(f"Death time: {sample_agent['death_time']}")
            print(f"Status: {'Alive' if sample_agent['death_time'] is None else 'Dead'}")
            print()
            
            # Get complete agent lifecycle
            print("1. Complete Agent Lifecycle:")
            try:
                lifecycle_result = get_lifecycle_tool.execute(
                    simulation_id=simulation_id,
                    agent_id=agent_id,
                    include_actions=True,
                    include_states=True,
                    include_health=True
                )
                
                if 'error' not in lifecycle_result:
                    agent_info = lifecycle_result.get('agent_info', {})
                    print(f"Lifespan: {agent_info.get('lifespan', 'N/A')} steps")
                    print(f"Initial resources: {agent_info.get('initial_resources', 'N/A')}")
                    print(f"Starting health: {agent_info.get('starting_health', 'N/A')}")
                    print(f"Genome ID: {agent_info.get('genome_id', 'N/A')}")
                    
                    # Show state count
                    state_count = lifecycle_result.get('state_count', 0)
                    print(f"State records: {state_count}")
                    
                    # Show action count
                    action_count = lifecycle_result.get('action_count', 0)
                    print(f"Action records: {action_count}")
                    
                    # Show health incidents
                    health_count = lifecycle_result.get('health_incident_count', 0)
                    print(f"Health incidents: {health_count}")
                    
                    # Show sample states
                    states = lifecycle_result.get('states', [])
                    if states:
                        print("\nSample states:")
                        for i, state in enumerate(states[:3]):
                            print(f"  Step {state['step']}: pos=({state['position']['x']:.1f}, {state['position']['y']:.1f}), resources={state['resources']:.1f}, health={state['health']}")
                    
                    # Show sample actions
                    actions = lifecycle_result.get('actions', [])
                    if actions:
                        print("\nSample actions:")
                        for i, action in enumerate(actions[:3]):
                            print(f"  Step {action['step']}: {action['action_type']} (reward: {action['reward']})")
                else:
                    print(f"Error in lifecycle analysis: {lifecycle_result['error']}")
                    
            except Exception as e:
                print(f"Error getting agent lifecycle: {e}")
            print()
            
            # Build agent lineage
            print("2. Agent Lineage:")
            try:
                lineage_result = build_lineage_tool.execute(
                    simulation_id=simulation_id,
                    agent_id=agent_id,
                    depth=2
                )
                
                if 'error' not in lineage_result:
                    agent = lineage_result.get('agent', {})
                    print(f"Agent: {agent.get('agent_id', 'N/A')} (Gen {agent.get('generation', 'N/A')})")
                    
                    ancestors = lineage_result.get('ancestors', [])
                    print(f"Ancestors: {len(ancestors)}")
                    for ancestor in ancestors:
                        print(f"  Parent: {ancestor.get('agent_id', 'N/A')} (Gen {ancestor.get('generation', 'N/A')})")
                    
                    descendants = lineage_result.get('descendants', [])
                    print(f"Descendants: {len(descendants)}")
                    for descendant in descendants:
                        print(f"  Child: {descendant.get('agent_id', 'N/A')} (Gen {descendant.get('generation', 'N/A')})")
                else:
                    print(f"Error in lineage analysis: {lineage_result['error']}")
                    
            except Exception as e:
                print(f"Error building agent lineage: {e}")
            print()
        else:
            print("No agents found for analysis")
            
    except Exception as e:
        print(f"Error getting agents for analysis: {e}")
    print()
else:
    print("No simulation ID available for agent analysis")
```
